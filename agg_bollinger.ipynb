### Aggressive Bollinger Bands Overview

Aggressive Bollinger Bands are a variation of the traditional Bollinger Bands indicator, designed to provide more immediate trading signals by adjusting the standard deviation (SD) multiplier. Bollinger Bands itself consists of a middle band (usually a simple moving average), an upper band (calculated as the middle band plus a multiple of the standard deviation), and a lower band (calculated as the middle band minus a multiple of the standard deviation).

#### Formula for Aggressive Bollinger Bands

The formula for calculating Aggressive Bollinger Bands typically involves using a smaller standard deviation multiplier compared to traditional Bollinger Bands. This adjustment makes the bands narrower and more responsive to price changes.

- **Middle Band (MB)**: Usually a simple moving average (SMA) over \( N \) periods.
  
- **Upper Band (UB)**: Calculated as \( \text{UB} = \text{MB} + k \cdot \sigma \), where \( k \) is a multiplier (typically less than 2) and \( \sigma \) is the standard deviation of prices over \( N \) periods.
  
- **Lower Band (LB)**: Calculated as \( \text{LB} = \text{MB} - k \cdot \sigma \).

### Example Calculation

Let's calculate Aggressive Bollinger Bands for a given set of prices.

#### Data
| Day | Price (P) |
|-----|-----------|
|  1  |    100    |
|  2  |    102    |
|  3  |     98    |
|  4  |    105    |
|  5  |    103    |
|  6  |    107    |
|  7  |    101    |

#### Step-by-Step Calculation

1. **Calculate Middle Band (MB)**:
   - Let's use a 5-period Simple Moving Average (SMA) for MB.
   - MB = \( \frac{100 + 102 + 98 + 105 + 103}{5} = 101.6 \)

2. **Calculate Standard Deviation (SD)**:
   - Calculate the standard deviation of prices over the last 5 periods.
   - SD = \( \sqrt{\frac{1}{5-1} \sum_{i=1}^{5} (P_i - \text{MB})^2} \)

3. **Calculate Aggressive Bollinger Bands**:
   - Choose a multiplier \( k \) (e.g., 1.0 for aggressive bands).
   - UB = MB + \( k \cdot \text{SD} \)
   - LB = MB - \( k \cdot \text{SD} \)

### Plotting the Aggressive Bollinger Bands

To visualize how Aggressive Bollinger Bands react to changes in price, let's plot the original prices and the bands using Python and Matplotlib.

#### Python Code Example

```python
import numpy as np
import matplotlib.pyplot as plt

# Data
days = np.arange(1, 8)
prices = np.array([100, 102, 98, 105, 103, 107, 101])

# Calculate MB and SD
N = 5
MB = np.zeros(len(prices) - N + 1)
SD = np.zeros(len(prices) - N + 1)

for i in range(len(MB)):
    MB[i] = np.mean(prices[i:i+N])
    SD[i] = np.std(prices[i:i+N], ddof=1)

# Calculate Aggressive Bollinger Bands
k = 1.0  # Aggressive multiplier
UB = MB + k * SD
LB = MB - k * SD

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(days, prices, marker='o', label='Price')
plt.plot(days[N-1:], MB, label='Middle Band (SMA)')
plt.plot(days[N-1:], UB, linestyle='--', color='g', label='Aggressive Upper Band')
plt.plot(days[N-1:], LB, linestyle='--', color='r', label='Aggressive Lower Band')
plt.title('Aggressive Bollinger Bands')
plt.xlabel('Day')
plt.ylabel('Price')
plt.legend()
plt.grid(True)
plt.show()
```

### Key Points to Remember

1. **Aggressive Multiplier**: Adjusts the sensitivity of the bands to price changes. A smaller multiplier (e.g., 1.0) creates narrower bands and more frequent signals.
2. **Immediate Signals**: Aggressive Bollinger Bands generate more immediate trading signals compared to traditional Bollinger Bands.
3. **Use Cases**: Useful for short-term traders looking for quick entries and exits based on price volatility.

### Advantages of Aggressive Bollinger Bands

1. **Quick Response**: Provides faster signals in response to price movements.
2. **Suitable for Short-Term Trading**: Ideal for traders focusing on short-term price volatility.
3. **Clearer Signals**: Narrower bands reduce ambiguity in interpreting signals.

### Disadvantages of Aggressive Bollinger Bands

1. **Increased Noise**: Higher sensitivity can lead to more false signals during choppy or ranging markets.
2. **Less Smooth**: The narrower bands may not capture broader trends as effectively as traditional Bollinger Bands.
3. **Parameter Sensitivity**: Choosing the right multiplier \( k \) can be critical and varies depending on market conditions.

### Summary

Aggressive Bollinger Bands offer a more responsive alternative to traditional Bollinger Bands, adjusting the standard deviation multiplier to create narrower bands and generate quicker trading signals. While they excel in providing immediate insights into price volatility, they also come with increased sensitivity and potential for noise, requiring careful parameter selection for effective use in trading strategies.


Aggressive Bollinger Bands can be used as signals in trading strategies to identify potential entry or exit points based on price volatility. Here’s how you can interpret and use these signals effectively:

### Entry Signals

1. **Bounce off Bands**:
   - **Long Entry**: When the price touches or crosses below the lower band and then bounces back up, it may signal a buying opportunity. This suggests that the price might have reached a temporary oversold condition.
   - **Short Entry**: Conversely, when the price touches or crosses above the upper band and then retreats, it may indicate a selling opportunity. This suggests that the price might have reached a temporary overbought condition.

2. **Band Squeeze**:
   - When the bands narrow significantly (due to decreased volatility), it indicates a period of consolidation. Traders often anticipate a breakout when volatility expands again. A long entry could be signaled when the price breaks above the upper band after a squeeze, and a short entry when it breaks below the lower band.

### Exit Signals

1. **Touch of Middle Band**:
   - As the price reverts to the middle band (SMA), it can signal an exit from a position. This is particularly useful for short-term traders who aim to capture swings within the bands.

2. **Band Expansion**:
   - After a period of consolidation (narrow bands), an expansion of the bands can signal renewed volatility and potentially a continuation of the trend. Traders may consider exiting a position if the price moves against their direction after such an expansion.

### Confirmation and Filter Signals

1. **Volume Confirmation**:
   - High trading volume accompanying a bounce off the bands can provide additional confirmation of the signal’s strength.

2. **Other Indicators**:
   - Combine Aggressive Bollinger Bands with other technical indicators (e.g., RSI, MACD) to confirm signals and reduce false positives.

### Example Strategy

Let's outline a simple trading strategy using Aggressive Bollinger Bands:

- **Long Entry**: Buy when the price touches or falls below the lower band and then rebounds above it.
  
- **Short Entry**: Sell when the price touches or rises above the upper band and then retreats below it.
  
- **Exit**: Exit the position when the price touches the middle band or when the bands expand significantly after a period of consolidation.

### Considerations

- **Market Conditions**: Adapt the multiplier \( k \) (e.g., 1.0 for aggressive bands) based on market volatility. Higher volatility may require a smaller \( k \) to avoid excessive signals.
  
- **Timeframe**: Adjust the period \( N \) of the SMA and the bands to suit your trading timeframe (e.g., shorter periods for intraday trading, longer periods for swing trading).

### Implementation

Implementing this strategy involves monitoring price movements relative to the Aggressive Bollinger Bands and making trading decisions based on the identified signals. Backtesting and practicing with historical data can help refine the strategy and optimize parameters for better performance in different market conditions.

````python
import datetime
import pytz
import pandas as pd
import MetaTrader5 as mt5
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
````

````python
#define assets
assets = ['EURUSD' ,'USDCHF' ,'GBPUSD' ,'USDCAD' ,'BTCUSD' ,'ETHUSD' ,'XAUUSD' ,'XAUSD' ,'SP500m','UK100']
frame15 = mt5.TIMEFRAME_M15
frame30 = mt5.TIMEFRAME_M30
frame1h = mt5.TIMEFRAME_H1
frame_d1 = mt5.TIMEFRAME_D1

timezone = pytz.timezone("Europe/Paris")
time_from = datetime.datetime(2013,6,24,tzinfo=timezone)
time_to = datetime.datetime(2024,6,24,tzinfo=timezone)

def get_assetsquotes(time_frame,year,month,day,asset):
    if not mt5.initialize():
        print("Initialize failed error code =",mt5.last_error())
        quit()
    timezone = pytz.timezone("Europe/Paris")
    time_from = datetime.datetime(year,month,day,tzinfo=timezone)
    time_to = datetime.datetime(2024,6,24,tzinfo=timezone)
    #time_to = datetime.datetime.now(timezone)+datetime.timedelta(days=1)
    rates = mt5.copy_rates_range(asset, time_frame, time_from,time_to)
    rates_frame = pd.DataFrame(rates)
    return rates_frame

def mass_import(asset,time_frame):
    if time_frame == 'H1':
        data = get_assetsquotes(frame1h,2013,2,2,assets[asset])
        data.iloc[:, 1:5].values
        data.round(decimals = 5)

    if time_frame == 'D1':
        data = get_assetsquotes(frame_d1,2000,1,1,asset = assets[asset])
        data.iloc[:, 1:5].values
        data.round(decimals = 5)
    return data

def delete_column(data, index, times):
    for i in range(1, times + 1):
        data = np.delete(data, index, axis = 1)
    return data

def add_column(data,times):
    for i in range(1,times+1):
        new = np.zeros((len(data),1),dtype=float)
        data = np.append(data,new,axis=1)
    return data

def ohlc_plot_bars(Data, window):
     
    Chosen = Data[-window:, ]
    
    for i in range(len(Chosen)):
        
        plt.vlines(x = i, ymin = Chosen[i, 2], ymax = Chosen[i, 1], color = 'black', linewidth = 1)  
        plt.vlines(x = i, ymin = Chosen[i, 2], ymax = Chosen[i, 1], color = 'black', linewidth = 1)
        
        if Chosen[i, 3] > Chosen[i, 0]:
            plt.vlines(x = i, ymin = Chosen[i, 0], ymax = Chosen[i, 3], color = 'black', linewidth = 1.00)  

        if Chosen[i, 3] < Chosen[i, 0]:
            plt.vlines(x = i, ymin = Chosen[i, 3], ymax = Chosen[i, 0], color = 'black', linewidth = 1.00)  
            
        if Chosen[i, 3] == Chosen[i, 0]:
            plt.vlines(x = i, ymin = Chosen[i, 3], ymax = Chosen[i, 0], color = 'black', linewidth = 1.00)  
            
    plt.grid() 


  
def jump(Data, jump):
    
    Data = Data[jump:, ]
    
    return Data

def ma(Data,lookback,close,where):
    Data = add_column(Data,1)

    for i in range(len(Data)):
        try:
            Data[i,where] = (Data[i-lookback+1:i+1,close].mean())

        except IndexError:
            pass
    
    Data = jump(Data,lookback)

    return Data

def delete_row(data, number):
    
    data = data[number:, ]
    
    return data

def volatility(data, lookback, close, position):
    
    data = add_column(data, 1)
    
    for i in range(len(data)):
        
        try:
            
            data[i, position] = (data[i - lookback + 1:i + 1, close].std())
    
        except IndexError:
            
            pass
     
    data = delete_row(data, lookback)    
     
    return data

def bollingerbands(Data,lookback,standard_deviation,close,position):
    data = add_column(Data,2)

    data = ma(data,lookback,close,position)

    data = volatility(data,lookback,close,position+1)

    data[:, position + 2] = data[:, position] + (standard_deviation * data[:, position + 1])
    data[:, position + 3] = data[:, position] - (standard_deviation * data[:, position + 1])
    
    data = delete_row(data, lookback)
    
    data = delete_column(data, position + 1, 1)
        
    return data

def signal(data, close, upper_boll, lower_boll, buy, sell):
 
    data = add_column(data, 10)
    
    for i in range(len(data)):  
        
        try:
        
            # Bullish signal
            if data[i, close] < data[i, lower_boll] and data[i - 1, close] > data[i - 1, lower_boll]:
                
                data[i + 1, buy] = 1
                     
            # Bearish signal
            elif data[i, close] > data[i, upper_boll] and data[i - 1, close] < data[i - 1, upper_boll]:
                
                data[i + 1, sell] = -1
                
        except IndexError:
                 
            pass
         
    return data

def ohlc_plot_candles(data, window):
      
    sample = data[-window:, ]
    
    for i in range(len(sample)):
        
        plt.vlines(x = i, ymin = sample[i, 2], ymax = sample[i, 1], color = 'black', linewidth = 1)  
        
        if sample[i, 3] > sample[i, 0]:
            
            plt.vlines(x = i, ymin = sample[i, 0], ymax = sample[i, 3], color = 'green', linewidth = 3)  

        if sample[i, 3] < sample[i, 0]:
            
            plt.vlines(x = i, ymin = sample[i, 3], ymax = sample[i, 0], color = 'red', linewidth = 3)  
            
        if sample[i, 3] == sample[i, 0]:
            
            plt.vlines(x = i, ymin = sample[i, 3], ymax = sample[i, 0] + 0.00003, color = 'black', linewidth = 1.00)  
            
    plt.grid()   

def signal_chart(data, position, buy_column, sell_column, genre = 'bars', window = 500):   
 
    sample = data[-window:, ]
    
    if genre == 'bars':

        fig, ax = plt.subplots(figsize = (10, 5))
        
        ohlc_plot_bars(data, window)    
    
        for i in range(len(sample)):
            
            if sample[i, buy_column] == 1:
                
                x = i
                y = sample[i, position]
            
                ax.annotate(' ', xy = (x, y), 
                            arrowprops = dict(width = 9, headlength = 11, headwidth = 11, facecolor = 'green', color = 'green'))
            
            elif sample[i, sell_column] == -1:
                
                x = i
                y = sample[i, position]
            
                ax.annotate(' ', xy = (x, y), 
                            arrowprops = dict(width = 9, headlength = -11, headwidth = -11, facecolor = 'red', color = 'red'))  

    if genre == 'candles':

        fig, ax = plt.subplots(figsize = (10, 5))
        
        ohlc_plot_candles(data, window)    
    
        for i in range(len(sample)):
            
            if sample[i, buy_column] == 1:
                
                x = i
                y = sample[i, position]
            
                ax.annotate(' ', xy = (x, y), 
                            arrowprops = dict(width = 9, headlength = 11, headwidth = 11, facecolor = 'green', color = 'green'))
            
            elif sample[i, sell_column] == -1:
                
                x = i
                y = sample[i, position]
            
                ax.annotate(' ', xy = (x, y), 
                            arrowprops = dict(width = 9, headlength = -11, headwidth = -11, facecolor = 'red', color = 'red'))  


````

````python
def mean(a,b):
    c = []
    if b == 'row':
        for i in range(len(a)):
            c.append(np.mean(a[i]))
    elif b == 'column':
        a = np.array(a)
        columns = [a[:, i] for i in range(a.shape[1])]
        for col in columns:
            c.append(np.mean(col))
        
    return c
````
