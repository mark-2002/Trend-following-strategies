### Exponential Moving Average (EMA) Overview

The Exponential Moving Average (EMA) is a type of weighted moving average that gives more significance to recent data points. This makes it more responsive to recent price changes than the Simple Moving Average (SMA), which assigns equal weight to all observations in the period.

#### Formula for EMA

The formula for calculating the EMA at time \( t \) (denoted as \( \text{EMA}_t \)) is:

\[ \text{EMA}_t = \alpha \cdot P_t + (1 - \alpha) \cdot \text{EMA}_{t-1} \]

Where:
- \( P_t \) is the price (or data point) at time \( t \).
- \( \alpha \) is the smoothing factor, calculated as \( \alpha = \frac{2}{N+1} \).
- \( N \) is the number of periods.

The initial EMA value (\( \text{EMA}_0 \)) is typically set to the first data point (or the SMA of the first \( N \) data points).

#### Smoothing Factor (Î±)

The smoothing factor \( \alpha \) determines the weight of the most recent price. It is derived from the number of periods \( N \) over which the EMA is calculated.

\[ \alpha = \frac{2}{N+1} \]

For example:
- For a 10-period EMA: \( \alpha = \frac{2}{10+1} = \frac{2}{11} \approx 0.1818 \)
- For a 20-period EMA: \( \alpha = \frac{2}{20+1} = \frac{2}{21} \approx 0.0952 \)

### Similarities Between EMA and Kaufman Adaptive Moving Average (KAMA)

1. **Trend Following**: Both EMA and KAMA are trend-following indicators, designed to smooth price data and identify the direction of the trend.
2. **Weighting Recent Data**: Both give more weight to recent data points, though they do so in different ways.
3. **Technical Analysis**: Both are used extensively in technical analysis for generating trading signals, identifying trends, and analyzing market momentum.

### Differences Between EMA and KAMA

1. **Calculation Method**:
   - **EMA**: Uses a constant smoothing factor \( \alpha \) calculated from a fixed period.
   - **KAMA**: Uses a variable smoothing factor that adjusts based on market volatility. It adapts the smoothing factor depending on the efficiency ratio (ER), which measures the trend's strength.

2. **Responsiveness**:
   - **EMA**: Has a fixed level of responsiveness determined by the period chosen.
   - **KAMA**: Adjusts its responsiveness based on the price action, becoming more responsive during trends and less responsive during sideways markets.

3. **Complexity**:
   - **EMA**: Simpler and easier to calculate.
   - **KAMA**: More complex due to the adaptive nature of its smoothing factor.

### Advantages of EMA Over KAMA

1. **Simplicity**: The EMA is straightforward to calculate and implement.
2. **Lower Computational Cost**: Requires less computational power and is easier to understand for beginners.
3. **Consistency**: Provides consistent results as it does not adapt to market conditions, which can be beneficial for some trading strategies.

### Advantages of KAMA Over EMA

1. **Adaptability**: KAMA adjusts its smoothing factor based on market conditions, making it more effective in different market environments.
2. **Noise Reduction**: By adapting to market volatility, KAMA can filter out more noise during sideways markets, providing clearer signals.
3. **Trend Sensitivity**: KAMA is more sensitive to trends, providing more accurate signals in trending markets.

### Disadvantages of EMA Over KAMA

1. **Lag**: EMA can lag behind the price, especially during sudden market movements.
2. **Less Adaptable**: EMA does not adjust to changing market conditions, which can lead to false signals in volatile or sideways markets.

### Disadvantages of KAMA Over EMA

1. **Complexity**: KAMA is more complex to calculate and understand, which may be a barrier for some traders.
2. **Higher Computational Cost**: Requires more computational power and can be slower to calculate compared to EMA.
3. **Parameter Sensitivity**: Choosing the right parameters for KAMA can be more challenging, and incorrect parameters can lead to suboptimal performance.

#### KAMA Calculation Example

KAMA involves more steps, including calculating the Efficiency Ratio (ER) and the smoothing factor.

1. **Efficiency Ratio (ER)**:
   - ER = (Change over period) / (Sum of absolute changes over each day in the period)
   - Change over period: \( P_t - P_{t-N} \)
   - Sum of absolute changes: \( \sum_{i=t-N}^{t} |P_i - P_{i-1}| \)

2. **Smoothing Constant**:
   - Fast EMA constant: \( \alpha_{fast} = \frac{2}{2+1} = 0.6667 \)
   - Slow EMA constant: \( \alpha_{slow} = \frac{2}{30+1} = 0.0645 \)
   - Smoothing constant \( \alpha = [ER \cdot (\alpha_{fast} - \alpha_{slow}) + \alpha_{slow}]^2 \)

3. **KAMA Calculation**:
   - KAMA_t = KAMA_{t-1} + \( \alpha \cdot (P_t - KAMA_{t-1}) \)

The smoothing factor (alpha) in the context of an EMA determines how much weight is given to the most recent data point versus the previous EMA value. The general formula for the smoothing factor (alpha) in an EMA is:

\[ \alpha = \frac{2}{N + 1} \]

where \( N \) is the lookback period.

````python
#define assets
assets = ['EURUSD' ,'USDCHF' ,'GBPUSD' ,'USDCAD' ,'BTCUSD' ,'ETHUSD' ,'XAUUSD' ,'XAUSD' ,'SP500m','UK100']
frame15 = mt5.TIMEFRAME_M15
frame30 = mt5.TIMEFRAME_M30
frame1h = mt5.TIMEFRAME_H1
frame_d1 = mt5.TIMEFRAME_D1

timezone = pytz.timezone("Europe/Paris")
time_from = datetime.datetime(2013,6,24,tzinfo=timezone)
time_to = datetime.datetime(2024,6,24,tzinfo=timezone)

def get_assetsquotes(time_frame,year,month,day,asset):
    if not mt5.initialize():
        print("Initialize failed error code =",mt5.last_error())
        quit()
    timezone = pytz.timezone("Europe/Paris")
    time_from = datetime.datetime(year,month,day,tzinfo=timezone)
    time_to = datetime.datetime(2024,6,24,tzinfo=timezone)
    #time_to = datetime.datetime.now(timezone)+datetime.timedelta(days=1)
    rates = mt5.copy_rates_range(asset, time_frame, time_from,time_to)
    rates_frame = pd.DataFrame(rates)
    return rates_frame

def mass_import(asset,time_frame):
    if time_frame == 'H1':
        data = get_assetsquotes(frame1h,2013,2,2,assets[asset])
        data.iloc[:, 1:5].values
        data.round(decimals = 5)

    if time_frame == 'D1':
        data = get_assetsquotes(frame_d1,2000,1,1,asset = assets[asset])
        data.iloc[:, 1:5].values
        data.round(decimals = 5)
    return data

def delete_column(data, index, times):
    for i in range(1, times + 1):
        data = np.delete(data, index, axis = 1)
    return data

def add_column(data,times):
    for i in range(1,times+1):
        new = np.zeros((len(data),1),dtype=float)
        data = np.append(data,new,axis=1)
    return data

def ohlc_plot_bars(Data, window):
     
    Chosen = Data[-window:, ]
    
    for i in range(len(Chosen)):
        
        plt.vlines(x = i, ymin = Chosen[i, 2], ymax = Chosen[i, 1], color = 'black', linewidth = 1)  
        plt.vlines(x = i, ymin = Chosen[i, 2], ymax = Chosen[i, 1], color = 'black', linewidth = 1)
        
        if Chosen[i, 3] > Chosen[i, 0]:
            plt.vlines(x = i, ymin = Chosen[i, 0], ymax = Chosen[i, 3], color = 'black', linewidth = 1.00)  

        if Chosen[i, 3] < Chosen[i, 0]:
            plt.vlines(x = i, ymin = Chosen[i, 3], ymax = Chosen[i, 0], color = 'black', linewidth = 1.00)  
            
        if Chosen[i, 3] == Chosen[i, 0]:
            plt.vlines(x = i, ymin = Chosen[i, 3], ymax = Chosen[i, 0], color = 'black', linewidth = 1.00)  
            
    plt.grid() 


  
def jump(Data, jump):
    
    Data = Data[jump:, ]
    
    return Data

def ma(Data,lookback,close,where):
    Data = add_column(Data,1)

    for i in range(len(Data)):
        try:
            Data[i,where] = (Data[i-lookback+1:i+1,close].mean())

        except IndexError:
            pass
    
    Data = jump(Data,lookback)

    return Data

def ema(Data,alpha,lookback,what,where):
    alpha = alpha/(lookback+1)
    beta = 1 - alpha

    Data = ma(Data,lookback,what,where)

    Data[lookback+1,where] = (Data[lookback+1,what]*alpha)+(Data[lookback,where]*beta)

    for i in range(lookback+2,len(Data)):
        try:
            Data[i,where] = (Data[i,what]*alpha) + (Data[i-1,where]*beta)

        except IndexError:
            pass

    return Data
````

````python
df = mass_import(0,'H1')
my_data1 = delete_column(df, 5, 3)
my_data1 = delete_column(my_data1,0,1)
````

````python
lookback = 100

my_data2 = ema(df,2,lookback=lookback,what=3,where=4)
````

````python
ohlc_plot_bars(my_data1, 500)
plt.plot(my_data2[-500:, 4], label = '100-period Exponential Moving Average')
plt.legend()
````
